```python
# Lab - 3
# Question - 1

import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import statistics as st
import sympy as sy
import math

data=pd.read_excel('ex2data1-logistic.xls')

x=np.array(data['x1'])
y=np.array(data['x2'])
y1=np.array(data['y'])
x=(x-np.mean(x))/np.std(x)
y=(y-np.mean(y))/np.std(y)
o0=10
o1=10
o2=10
x00 = o0
x11 = o1
x22 = o2

a1 = []
for i in range(100):
  a1.append(i);
    
data1 = random.sample(a1,k=90);   # take k length list of unique elements.
data1.sort();




prev0=o0+1
prev1=o1+1
prev2=o2+1

eps=1e-6
alpha=0.01
ni=0

while(abs(o0-prev0)>=eps or abs(o1-prev1)>=eps or abs(o2-prev2)>=eps):
    prev0=x00
    prev1=x11
    prev2=x22
    #alpha=alpha-diff1(alpha,prev0,prev1)/diff2(alpha,prev0,prev1)
    
    #ans = 1/(1+math.exp((-1)*sum([(prev0+prev1*x[i]+prev2*y[i]) for i in range(len(x))])));
    #alpha*(sum([(Y[arr[i]] - ( 1.0 /  (1 + math.exp(-t0-t1*X1[arr[i]] - t2*X2[arr[i]]) ) )) for i in range(len(arr))]))
    o0 = x00 + alpha*(sum([(y1[data1[i]]- (1.0/(1 + math.exp(-prev0-prev1*x[data1[i]]-prev2*y[data1[i]])))) for i in range(len(data1))]))
    o1 = x11 + alpha*(sum([((y1[data1[i]]- (1.0/(1 + math.exp(-prev0-prev1*x[data1[i]]-prev2*y[data1[i]]))))*x[data1[i]]) for i in range(len(data1))]))
    o2 = x22 + alpha*(sum([((y1[data1[i]]- (1.0/(1 + math.exp(-prev0-prev1*x[data1[i]]-prev2*y[data1[i]]))))*y[data1[i]]) for i in range(len(data1))]))
    
    x00 = o0
    x11 = o1
    x22 = o2
    ni=ni+1
    #print(o0,o1,o2)

count = 0;
for i in range(100):
   if(i not in  data1):
        if( (1.0 / (1 + math.exp(-x00-x11*x[i] - x22*y[i]) ) ) >= 0.5):
            if(y1[i]!=1):
                count = count + 1;
        else:
            if(y1[i]!=0):
                count = count + 1;
    

plt.scatter(x,y)

for i in range(len(x)):
    if(y1[i] == 1):
        plt.scatter(x[i],y[i],c="red")
    else:
        plt.scatter(x[i],y[i],c="Blue")
        

print('Theta-0 : ',x00)
print('Theta-1 : ',x11)
print('Theta-2 : ',x22)
print('Number Of Iteration : ' , ni)
print('Error in Testing Data : ',count)
print('Accuracy',1-count/10)

xi = np.linspace(-2.0, 2.0, 100)
yi = np.linspace(-2.0, 2.0, 100)
X, Y = np.meshgrid(xi,yi)  # creates rectangular grid of (xi,yi) . X,Y= X matrix and Y matrix.
F = x00 + x11*X + x22*Y
plt.contour(X,Y,F,[0])
plt.show()



```
