```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns;
import statistics as st
import sympy as sy
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.cluster import KMeans 
from sklearn import metrics 
from scipy.spatial.distance import cdist
```

```python
def kmeans(X,no_clusters):
    ind=np.random.randint(low=0, high=X.shape[0], size=no_clusters)
    centroids=X[ind]
    
    while True:
        clusters = pairwise_distances_argmin_min(X,centroids)
        clusters = clusters[0]
        
        new_centroids=np.array([X[clusters==i].mean(0) for i in range(no_clusters)]);
        
        if(np.all(centroids==new_centroids)):
            break;
        else:
            centroids=new_centroids
            
            
    return new_centroids, clusters
    
```

```python
#dataset 1

a=pd.read_excel(r'C:\Users\LENOVO\Desktop\kmeans_data.xlsx')
x=np.array(a['x'])
y=np.array(a['y'])
X=np.array([x,y])
X=X.T

plt.plot()
plt.scatter(x,y)
plt.show()

plt.plot()
cen,clu=kmeans(X,1)
plt.title('k=1')
plt.scatter(x,y,c=clu)
plt.scatter(cen[:,0],cen[:,1],c='black',s=200)
plt.show()

plt.plot()
cen,clu=kmeans(X,2)
plt.title('k=2')
plt.scatter(x,y,c=clu)
plt.scatter(cen[:,0],cen[:,1],c='black',s=200)
plt.show()

plt.plot()
cen,clu=kmeans(X,3)
plt.title('k=3')
plt.scatter(x,y,c=clu)
plt.scatter(cen[:,0],cen[:,1],c='black',s=200)
plt.show()

plt.plot()
cen,clu=kmeans(X,4)
plt.title('k=4')
plt.scatter(x,y,c=clu)
plt.scatter(cen[:,0],cen[:,1],c='black',s=200)
plt.show()

#elbow method
wss = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(k).fit(X)
    kmeanModel.fit(X)
    wss.append(kmeanModel.inertia_)
    
plt.plot(K, wss, 'bx-')
plt.xlabel('k')
plt.ylabel('WSS')
plt.show()
```
